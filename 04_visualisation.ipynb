```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

path = pd.read_csv('titanic/train.csv')

sns.set(style='darkgrid')
```

```python
sns.histplot(path['Age'].dropna(), bins=40)
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()
```

```python
sns.boxplot(x='Sex', y='Age', data=path)
plt.title('Возраст по полу')
plt.show()
```

```python
path_sorted = path.sort_values(by='PassengerId')
sns.lineplot(x='PassengerId', y='Fare', data=path_sorted)
plt.title('Тариф по пассажирам')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=path)
plt.title('Возраст и тариф(по выживаемости)')
plt.show()
```

```python
numeric_path = path[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch', 'Survived']]
corr = numeric_path.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Корреляция между признаками')
plt.show()
```
